```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)
```

```python
cars = pd.read_csv('cars.csv')
```

```python
cars.info()
```

```python
cars = pd.read_csv('cars.csv').sample(n=15)
```

```python
c = cars.head()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=cars,x='Date Auctioned',y='Locale',hue = 'Manufacturer')
plt.title('Histogram of Date Auctioned by Locale and Manufacturer')
plt.xticks(rotation=45)
```

```python
cars = pd.read_csv('cars.csv').sample(n=15)
plt.figure(figsize=(12, 6))
sns.boxplot(data=cars, x='Manufacturer', y='Original Price',hue='Manufacturer', palette='Set3',legend=False)
plt.title('Price Distribution by Manufacturer')
plt.xticks(rotation=45)
plt.show()
```

```python
cars = pd.read_csv('cars.csv').sample(n=15)
plt.figure(figsize=(8, 3))
sns.countplot(data=cars, x='Manufacturer',hue='Manufacturer', palette='Pastel1',legend = False)
plt.title('Count of Auctioned Cars')
plt.xticks(rotation=45)
plt.show()
```

```python
g = sns.FacetGrid(cars, col='Manufacturer', hue='Model Year', col_wrap=3)
g.map(sns.scatterplot, 'Model Year', 'Date Auctioned')
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Scatterplot of Model Year vs. Date Auctioned by Manufacturer')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=cars, x='Original Price', y='Car Name', hue='Manufacturer', palette='cividis')
plt.title('Bar Plot of Original Price by Car Name')
plt.xlabel('Original Price')
plt.ylabel('Car Name')
plt.xticks(rotation=45)
plt.legend(title='Manufacturer', loc='lower left')
plt.tight_layout()
plt.show()
```
